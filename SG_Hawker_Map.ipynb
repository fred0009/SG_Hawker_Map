```python
import folium
import json
import textwrap
```

```python
#Load file
file_line = 'data/hawker-centres-geojson.geojson'

with open(file_line, 'r') as f:
    data = json.load(f)
```

```python
# Text extraction
for i in range(len(data['features'])):
    data['features'][i]['properties']['Hawker Name'] = data['features'][i]['properties']['Description'].split('<th>NAME</th> <td>')[1].split('</td> </tr><tr bgcolor')[0]
    data['features'][i]['properties']['Hawker Desc'] = data['features'][i]['properties']['Description'].split('DESCRIPTION_MYENV</th> <td>')[1].split('</td> </tr><tr bgcolor')[0]
```

```python
# Combine Hawker name and description
for i in range(len(data['features'])):
    data['features'][i]['properties']['Hawker Info'] = '<b>Hawker Name: </b>' + data['features'][i]['properties']['Hawker Name']+ '<br><br><b>Hawker Descr: </b>' + data['features'][i]['properties']['Hawker Desc']
```

```python
# Adding <br> (line break in html) to hawker description
for i in range(len(data['features'])):
    data['features'][i]['properties']['Hawker Info'] = "<br>".join(textwrap.wrap(data['features'][i]['properties']['Hawker Info'], width=60))
```

```python
# visualize and save

sg_map = folium.Map(location=[1.3521, 103.8198],
                   tiles='CartoDB positron ', zoom_start=12)
folium.GeoJson(data, 
               tooltip=folium.features.GeoJsonTooltip(
                    fields=['Hawker Info'], labels=False,
                    popup=False,
                   offset=(0, 0)
                ),
              ).add_to(sg_map)
sg_map.save(outfile= "output/test.html")
```
